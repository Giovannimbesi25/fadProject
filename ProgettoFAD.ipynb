# Drug Consumption Analysis <a class="anchor" id="title"></a>

<p>
Il database contiene record per 1885 intervistati. Per ogni intervistato sono noti 12 attributi: misurazioni della personalità che includono NEO-FFI-R (nevroticismo, estroversione, apertura all'esperienza, gradevolezza e coscienziosità), BIS-11 (impulsività) e ImpSS (ricerca di sensazioni), livello di istruzione, età, sesso, paese di residenza ed etnia.
Tutti gli attributi di input sono originariamente categorici e sono quantificati. Dopo la quantificazione, i valori di tutte le feature di input possono essere considerati come valori reali. Inoltre, i partecipanti sono stati interrogati sul loro uso di 18 droghe legali e illegali (alcol, anfetamine, nitrito di amile, benzodiazepine, cannabis, cioccolato, cocaina, caffeina, crack, ecstasy, eroina, ketamina, droghe legali, LSD, metadone, funghi, abuso di nicotina e sostanze volatili e un farmaco fittizio (Semeron), il quale è stato introdotto per identificare i record non validi.
Per ogni farmaco l'intervistato ha selezionato una delle risposte: non ha mai usato il farmaco, l'ha usato più di dieci anni fa o nell'ultimo decennio, anno, mese, settimana o giorno.

## Distribuzione dei dati <a class="anchor" id="distribuzione"></a>

<p>Si procede all'analisi della distribuzione dei dati sulle varie features</p>

```python
drugs_data.describe()
```

```python
fig, axes = plt.subplots(6,3,figsize = (25,15))
fig.suptitle("Count of Different Classes Vs Drug",fontsize=15)
k=0
for i in range(6):
    for j in range(3):
        if k < len(drugs):
            sns.countplot(x=drugs[k], data=drugs_data,ax=axes[i][j])
            k+=1
plt.tight_layout()
plt.show()
```

<p>In questo grafico è stata analizzata la distribuzione delle classi per ogni sostanza, evidenziando come, ad eccezzione di Alcol, Cioccolata, Caffeina e Nicotina, la maggior parte degli intervistati non abbiano mai usato queste droghe o perlomeno non in maniera considerevole. Un'interessante considerazione può essere fatta sulla Cannabis, la cui distribuzione è molto più uniforme rispetto a tutte le altre.</p>

```python
ethnicities = ['Asian','Black','Mixed-Black/Asian','Mixed-White/Asian','Mixed-White/Black','Other','White']
educations = ['Left school before 16y' ,'Left school at 16', 'Left school at 17', 'Left school at 18' ,'Some college or university',              'Prof. certificate/diploma', 'University degree', 'Masters degree', 'Doctorate degree']
ages = ['18-24','25-34','35-44','45-54','55-64','65+']
genders = ['M', 'F']

demographics = {'Education': educations, 'Ethnicity': ethnicities, 'Age': ages, 'Gender': genders}

def plot_drug_by_demographics(drug):
    fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(35, 6))
    sns.set_style('whitegrid')
    for i, (demographic, categories) in enumerate(demographics.items()):
        ax = axes.flatten()[i]
        sns.barplot(data=drugs_data, x=demographic, y=drug, ax=ax)
        ax.set(xlabel=demographic, ylabel=f'Frequency of {drug} Consumption')
        ax.set_xticklabels(labels=categories, rotation=20)
        ax.set_title(f'{demographic} vs {drug}', y=1.03, fontsize=16)
    plt.tight_layout()
    plt.show()


for drug in ['Alcohol', 'Cannabis', 'Nicotine']:
  plot_drug_by_demographics(drug)
```

<p>Un'altro aspetto interessante riguarda l'analisi dell'utilizzo delle varie sostanze sulla base dei fattori quali l'età, l'etnia, il genere ed il livello di istruzione. È interessante notare come, se per l'Alcol non si registra una grande varianza nei dati, Nicotina e Cannabis vedono una forte diminuzione del loro utilizzo col crescere dell'età ed un dislivello nel genere, a favore di quello maschile. Da sottolineare inoltre come in media il consumo di Cannabis risulta maggiore negli individui di razza bianca.</p>

## Analisi delle correlazioni <a class="anchor" id="correlazioni"></a>

<p>Si è deciso di utilizzare l'indice di Pearson per analizzare le correlazioni tra le varie features</p>

```python
corr = drugs_data.corr()
plt.figure(figsize=(30,10))
sns.heatmap(corr, annot=True, vmin=-1)
plt.title('Correlations', size=16)
plt.xlabel('Features', size=12)
plt.ylabel('Features', size=12)
plt.show()
```

<p>A colpo d'occhio è possibile notare come poche features risultano avere una forte correlazione, sia in positivo che in negativo. In generale, i valori più alti si registrano nella parte in basso a destra della matrice anche se in pochi casi superano la soglia minima dello 0.5.

Tenendo presente le considerazioni ottenute, si è deciso di sviluppare delle analisi incentrate nello studio dell'utilizzo di ciascuna sostanza, sulla base dei fattori della personalità.</p>

# Regressione <a class="anchor" id="regressione"></a>



## Utility <a class="anchor" id="reg-utils"></a>

```python
def preprocessing_inputs(df, column):
    y = df[column]
    X = df.drop(column, axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train), 
                           index=X_train.index, 
                           columns=X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), 
                          index=X_test.index, 
                          columns=X_test.columns)
    
    return X_train, X_test, y_train, y_test

```

<p>Questa funzione permette di normalizzare e separare il dataset in training e test set. Essa verrà utilizzata anche in future trattazioni.</p>

```python
def get_best_features(X, y, model):

  cv = KFold(n_splits=5, shuffle=True, random_state=42)
  rfecv = RFECV(estimator=model, step=1, cv=cv, scoring='neg_mean_absolute_error', n_jobs=-1)
  rfecv.fit(X, y)
  n_features = rfecv.n_features_
  rfe = RFE(estimator=model, n_features_to_select=n_features)
  rfe.fit(X, y)
  selected_features = X.columns[rfe.support_]

  return selected_features
```

<p>Tramite l'tulizzo della recursive feature elimination sono state stimate quante e quali feature utilizzare per la regressione. </p>

```python
def drugs_user_regression(drug, drugs_data, number_of_drugs):
    N = number_of_drugs + 1
    drugs_corr = drugs_data[drugs].corr()
    drug_corr = drugs_corr[drug].sort_values(ascending=False)
    drug_df = drugs_data[list(drug_corr.index[:N])]
    drug_df['User'] = drug_df.apply(lambda row: (row >= 3).sum(), axis=1)
    drug_df = drug_df.drop(list(drug_corr.index[:N]), axis=1)
    drug_df = pd.concat([drugs_data[personal_traits], drug_df], axis=1)
    X_train, X_test, y_train, y_test = preprocessing_inputs(drug_df, 'User')

    lr = LinearRegression()
    selected_features = get_best_features(X_train, y_train, lr)
    drug_df = drug_df[selected_features.union(['User'])]

    X_train, X_test, y_train, y_test = preprocessing_inputs(drug_df, 'User')
    lr.fit(X_train, y_train)

    y_pred = lr.predict(X_test)

    mse = mean_absolute_error(y_test, y_pred)

    rscore = r2_score(y_test, y_pred)

    return mse, rscore

def regression_evaluation(drugs, drugs_data, number_of_drugs):
    mae_scores = []
    r_scores = []
    for drug in drugs:
        mae, rscore = drugs_user_regression(drug, drugs_data, number_of_drugs)
        mae_scores.append(mae)
        r_scores.append(rscore)
    mean_mae = sum(mae_scores) / len(mae_scores)
    mean_rscore = sum(r_scores) / len(r_scores)
    return mean_mae, mean_rscore
    

N_values = [2, 3, 5, 7, 10, 15, (len(drugs)-1)]
mae_scores, r_scores = [], []
for N in N_values:
  mean_mae, mean_rscore = regression_evaluation(drugs, drugs_data, N)
  mae_scores.append(mean_mae)
  r_scores.append(mean_rscore)
```

La regressione lineare è basata sui seguenti passi: 
1. Selezionare una sostanza.
2. Considerare le N sostanze più correlate a quella in esame.
3. Definire una nuova classe che, per ogni intervistato, indichi il numero di droghe utilizzate di recente, tra quelle selezionate.
4. A partire dai tratti della personalità si selezionano le features più interessanti per predire il valore della classe sopra definita per ciascun partecipante.

# Classificazione <a class="anchor" id="classificazione"></a>


<p>In questo paragrafo verranno utilizzati una serie di algoritmi per la classificazione binaria
</p>

## Utility <a class="anchor" id="class-utils"></a>

```python
models = {
            'LogisticRegression': LogisticRegression(),
            'SupportVectorMachines': SVC(),
            'DecisionTreeClassifier': DecisionTreeClassifier(),
            'Random Forest Classifier': RandomForestClassifier()
}
```

<p>Quì sopra vengono elencati i modelli utilizzati per effettuare la classificazione</p>

```python
def findBestValues(name,model, X_train, y_train):
    if(name == 'LogisticRegression'):
      params = {'penalty': ['l1', 'l2'], 'C': [0.1, 1, 10, 100]}
    if(name == 'SupportVectorMachines'):
      params = {'C': [0.01, 0.1, 1, 10], 'kernel': ['linear', 'rbf', 'poly'], 'gamma': ['scale', 'auto']}
    if(name == 'DecisionTreeClassifier'):
      params = {'max_features': ['auto', 'sqrt', 'log2'],'max_depth' : [2,4],'criterion' :['gini', 'entropy']}
    if(name == 'Random Forest Classifier'):
      params = {'n_estimators': [50, 100 , 150],'max_depth': [2, 4],'min_samples_split': [2, 4],'min_samples_leaf': [1, 2],'max_features': ['sqrt', 'log2']}
    grid_model = GridSearchCV(model, params, cv=5)
    grid_model.fit(X_train, y_train)
    best_model = model.__class__(**grid_model.best_params_)
    return best_model
```

<p>Mediante l'utilizzo della GridSearch sono stati stimati i migliori parametri per ciascun modello</p>

```python
def compute_scores(X_train, y_train, X_test, y_test):
    scores = {'accuracy': {}, 'f1': {}, 'precision': {}, 'recall': {}}
    for name, model in models.items():
        ft_model = findBestValues(name, model, X_train, y_train)
        ft_model.fit(X_train, y_train)
        yhat = ft_model.predict(X_test)
        scores['accuracy'][name] = accuracy_score(y_test, yhat)
        scores['f1'][name] = f1_score(y_test, yhat, pos_label=1)
        scores['precision'][name] = precision_score(y_test, yhat, pos_label=1)
        scores['recall'][name] = recall_score(y_test, yhat, pos_label=1)
    return scores
```

<p>Le metriche utilizzate per la valutazione degli algoritmi di classificazione sono quattro: Accuracy, F1_score, Precision e Recall. In particolare le ultime due sono state utilizzate in quanto, analizzando le matrici di confusione, alcune sostanze presentano una discrepanza tra il numero di osservazioni classificate come positive e come negative.</p>

```python
def drug_user_classification(drug, drugs_data, show):
  print(drug + " in progress...")
  drug_df = pd.concat([drugs_data[personal_traits], drugs_data[drug]], axis=1)
  column_name = drug + '_User'
  drug_df[column_name] = drug_df[drug].apply(lambda x: 1 if x not in [0,1] else 0)
  drug_df = drug_df.drop([drug], axis=1)

  X_train, X_test, y_train, y_test = preprocessing_inputs(drug_df, column_name)
  scores = compute_scores(X_train, y_train, X_test, y_test)

  if(show == True):
    colors = ['C0', 'C1', 'C2', 'C3']
    axs = plt.subplots(nrows=2, ncols=2, figsize=(25,8))
    axs = axs.ravel()
    for i, metric in enumerate(['accuracy', 'f1', 'precision', 'recall']):
        axs[i].bar(scores[metric].keys(), scores[metric].values(), color=colors)
        axs[i].set_xlabel('Model', fontsize=14)
        axs[i].set_ylabel(metric.capitalize(), fontsize=14)
        axs[i].set_title(f'{metric.capitalize()} Scores for {drug}', fontsize=14)
        axs[i].tick_params(axis='x', labelrotation=0)
        for j, score in enumerate(scores[metric].values()):
            axs[i].text(j, score + 0.01, f'{score:.2f}', ha='center', va='bottom')

    plt.subplots_adjust(wspace=0.5, hspace=0.5)

    
    plt.tight_layout()
    plt.show()

  else:
    return scores
```

La classificazione viene effetuata seguendo questi step:
1. Scelta una sostanza, viene creata una nuova feature "User".
2. Se l'intervistato ha usato quella sostanza di recente, si assegna 1 alla nuova feature, altrimenti 0.
3. Si utilizzano i tratti personali dell'intervistato per effettuare la classificazione.

# Clustering <a class="anchor" id="clustering"></a>



## Algoritmi utilizzati

<p>Gli algoritmi utilizzati per effettuare il clustering sono il KMeans, DBScan ed il clustering Gerarchico.</p>

```python
algorithms = ['KMeans', 'DBScan', 'Hierarchical']
```

```python
def apply_clustering(data_cluster):
    cluster_info = {}
    sil_scores = {}

    scaler = StandardScaler()
    x_scaled = scaler.fit_transform(data_cluster)
    scoring_metric = make_scorer(silhouette_score)

    for alg in algorithms:
        if alg == 'KMeans':
            kmeans = KMeans()
            params = {'n_clusters': [2, 3, 4, 6, 8, 10], 'init': ['k-means++', 'random'], 'n_init': [10, 20, 30]}
            grid_search = GridSearchCV(kmeans, param_grid=params, scoring=scoring_metric, cv=5)
            clusters = grid_search.fit(x_scaled).best_estimator_.fit_predict(x_scaled)
            n_clusters = grid_search.best_params_['n_clusters']
            cluster_info[alg] = n_clusters

        elif alg == 'DBScan':
            dbscan = DBSCAN()
            params = {'eps': [2, 3, 4, 5], 'min_samples': [3, 4, 5], 'metric': ['euclidean', 'callable']}
            grid_search = GridSearchCV(dbscan, param_grid=params, scoring=scoring_metric, cv=5)
            clusters = grid_search.fit(x_scaled).best_estimator_.fit_predict(x_scaled)
            n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
            cluster_info[alg] = n_clusters

        elif alg == 'Hierarchical':
            hierarchical = AgglomerativeClustering()
            params = {'n_clusters': [2, 3, 4, 6, 8, 10], 'linkage': ['ward', 'complete', 'average'],
                      'affinity': ['euclidean', 'manhattan', 'cosine']}
            grid_search = GridSearchCV(hierarchical, param_grid=params, scoring=scoring_metric)
            clusters = grid_search.fit(x_scaled).best_estimator_.fit_predict(x_scaled)
            n_clusters = grid_search.best_params_['n_clusters']
            cluster_info[alg] = n_clusters

        data_cluster['clusters'] = clusters
        sil_scores[alg] = silhouette_score(x_scaled, data_cluster['clusters'], metric='euclidean')

    return cluster_info, sil_scores
```

<p>L'analisi prevede di utilizzare una GridSearch per definire i parametri migliori per ciascuno dei metodi di clustering.</p>

## Approccio Alternativo

Lo scopo di questa analisi è quello di definire una feature ausiliaria che indichi per ciascun individuo, l'utilizzo recente di almeno una sostanza illegale. L'idea è quella di applicare il clustering gerarchico sul dataset privato della nuova feature, e vedere come si dispongono gli elementi sulla base di questa nuova 'classe'.

```python
data = drugs_data.copy()
data['illegal_user'] = 0

for index, row in data.iterrows():
    count = 0
    for drug in illegal_drugs:
        if row[drug] >= 5:
            count += 1
            if count >= 1:
                data.at[index, 'illegal_user'] = 1
                break 
```

Una volta definita la nuova feature, si procede all'utilizzo dell'algoritmo di clustering gerarchico per separare i dati in due clusters. Tale numero è stato utilizzato in quanto ricorrente nella precedente analisi.

```python
ac = AgglomerativeClustering(n_clusters=2)
X = data.drop(['illegal_user'], axis = 1)

clusters = ac.fit_predict(X)

cluster_data=data
cluster_data["cluster"]=clusters
result = cluster_data.groupby(['cluster', 'illegal_user']).size().unstack(fill_value=0)
result = result.divide(result.sum(axis=1), axis=0)
```

### Risultati

Il clustering ha fornito dei risultati interessanti: le osservazioni di classe 1, ovvero quegli individui che hanno assunto di recente una sostanza illegale, sono stati inseriti quasi esclusivamente in un unico cluster. L'altro cluster individuato è invece composto per l'80% da individui che non hanno assunto di recente droghe illegali.
